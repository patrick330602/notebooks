### Import packages

```python
%matplotlib inline
import matplotlib.pyplot as plt
from sklearn.datasets import make_classification
```

### Make a dataset
Let's make a dataset that has/is
* 2 features
* 1 cluster per class
* well seperable


```python
X, y = make_classification(n_features=2, n_redundant=0, n_clusters_per_class=1, n_classes=2, class_sep=3.)
```

### What's the shape of our features/labels?

```python
print(X.shape, y.shape)
```

### map our labels to an interpretable color space

```python
color_dict = {0: 'r', 1: 'b'}
colors = [color_dict[i] for i in y]
```

### plot the features & labels

```python
plt.scatter(X[:, 0], X[:, 1], c=colors)
plt.show()
```
